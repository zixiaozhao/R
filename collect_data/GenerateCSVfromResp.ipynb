```python
import json
import requests
import csv
import os
```

```python
headers = {'Authorization': ''}
```

```python
# Getting GitHub Metadata and contentlinks to get file links

with open("ghMetaData.csv", "a+", newline="") as f1, open("contentsLinks.txt", "a", newline = "") as contentLinks:
    writer = csv.writer(f1)
    writer.writerow(["id", "repo", "html_url","owner_login", "contents_url", "created_at", "last_commit", "contributor_count","language", "star_count", "forks_count"])
    with open("jsonURLs.txt", "r") as file:
        for urls in file:
            response = requests.get(urls, headers=headers)
            
            if response.status_code == 200:
                data = response.json()
                for item in data['items']:
                    contributors_url = item.get("contributors_url")
                    contributors = requests.get(contributors_url, headers=headers)
                    if contributors.status_code == 200:
                         contributor = contributors.json()
                         contributor_count = len(contributor)
                    id = item.get("id")
                    repo = item.get("full_name")
                    html_url = item.get("html_url")
                    owner_login = item.get("owner", {}).get("login")
                    contents_url = item.get("contents_url")
                    contents_url = str(contents_url[:-7])
                    created_at = item.get("created_at")
                    last_commit = item.get("updated_at")
                    language = item.get("language")
                    stargazers_count = item.get("stargazers_count")
                    forks_count = item.get("forks_count")
                    writer.writerow([id, repo, html_url, owner_login, contents_url, created_at, last_commit, contributor_count, language, stargazers_count, forks_count])
                    contentLinks.write(contents_url)
            else:
                print("Failed to get data")
```

```python
with open("ghMetaData.csv", "r") as metaData, open("MetaData.csv", "a+", newline="") as f1:
    writer = csv.writer(f1)
    writer.writerow(["id", "repo", "html_url","owner_login", "contents_url", "created_at", "last_commit", "contributor_count","language", "star_count", "forks_count", "tidy", "# of R file"])
    reader = csv.reader(metaData)
    next(reader)
    for row in reader:
        des_file = "" + row[1] + "/DESCRIPTION.txt"
        des = "" + row[1]
        if not os.path.exists(des_file):
            continue
        files = os.listdir(des)
        num = len(files)-1
        with open(des_file, 'r') as file:
            line = file.readline()
            flag = False
            tidy = False
            while line:
                if line.__contains__(':'):
                    if line.__contains__('Suggests') or line.__contains__('Imports'):
                        flag = True
                    else:
                        flag = False
                elif line.__contains__('ggplot2') or line.__contains__('dplyr') or line.__contains__('tidyr') or line.__contains__('readr') or line.__contains__('purrr') or line.__contains__('tibble') or line.__contains__('stringr') or line.__contains__('forcats'):
                    if flag:
                        tidy = True
                line = file.readline()
        writer.writerow([row[0], row[1], row[2], row[3], row[4], row[5], row[6], row[7], row[8], row[9], row[10], tidy, num])

```

```python
with open("ghMetaData.csv", "r") as metaData:
    reader = csv.reader(metaData)
    next(reader)
    for row in reader:
        print(row[1])
        des_file = "" + row[1] + "/man"
        if os.path.exists(des_file):
            continue
        os.makedirs(des_file)

        url = row[4]+"man"
        d = requests.get(url, headers=headers)
        if d.status_code == 200:
            dd = d.json()
            for item in dd:
                if item['name'].endswith('.Rd'):
                    r = requests.get(item.get("download_url"))
                    path = des_file + "/" + item.get("name")
                    open(path, 'wb').write(r.content)
            
```

```python
with open("FileLinks.txt", "a", newline="") as fLinks, open("contentsLinks.txt", "r") as cLinks:
    count = 0
    for urls in cLinks:
        response = requests.get(urls[:-1], headers=headers)
        print(urls)
        
        #response = requests.get(urls, headers=headers)
        if response.status_code == 200:
            data = response.json()
            for file in data:
                if file['name'].endswith('.py'):
                    fLinks.write(str(file['download_url'])+ "\n")
                    print(file['download_url'])
                    count+=1
        else:
            print("Failed to get data")
        print(count)
```

```python
import os
import jsonlines
import csv
import random
import numpy as np
```

```python

data_path = "" + 'Rcombine/'
data = []

with open("MetaData.csv", "r") as metaData:
    reader = csv.reader(metaData)
    next(reader)
    for row in reader:
        if row[11] :
            print(row[11])
            des_file = "" + row[1] + '/functions.jsonl'
            if os.path.exists(des_file):
                with jsonlines.open(des_file) as reader:
                    for obj in reader:
                        data.append(obj)
random.shuffle(data)
indicies_for_splitting = [int(len(data) * 0.8), int(len(data) * (0.9))]
train, val, test = np.split(data, indicies_for_splitting)
with jsonlines.open(data_path+'train.jsonl', 'a') as tw, jsonlines.open(data_path+'valid.jsonl', 'a') as vw, jsonlines.open(data_path+'test.jsonl', 'a') as tew:
    for i in train:
        tw.write(i)
    for i in test:
        tew.write(i)
    for i in val:
        vw.write(i)
```
